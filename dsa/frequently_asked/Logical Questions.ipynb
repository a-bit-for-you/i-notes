# 6. Rotate array

```python
def rotate(arr):
    len_arr = len(arr)
    m = len_arr - 1
    for i in range(m):
        arr[i], arr[m] = arr[m], arr[i]
        print(arr)
    return arr
rotate([1,2,3,4,5])
```

### Find the sub array with given Sum

```python
from collections import defaultdict

def subArraySum(arr, s): 
    dic = defaultdict(int)
    start = 0
    end = -1
    current_sum = 0
    for i in range(len(arr)):
        current_sum += arr[i]

        if current_sum == s:
            return start, i

        if current_sum - s in dic:
            start = dic[current_sum - s] + 1
            end = i
            return start, end

        dic[current_sum] = i
    return (-1,)
subArraySum([1,2,3,7,5], 12)
```

## fibo using recursion

```python
def fibo(n):
    if n <= 2: 
        return 1
    else: 
        return fibo(n-1) + fibo(n-2)
    
for i in range(1,10):
    print(fibo(i))
```

## fibo using memoization

```python
memo = {}
def fibo(n):
    if n in memo:
        return memo[n]
    if n<=2:
        return 1
    else:
        memo[n] = fibo(n-1) + fibo(n-2)
        return memo[n]
        
for i in range(1, 10):
    print(fibo(i))
```

## fibo using memoization, (built-in method)

```python
from functools import lru_cache

@lru_cache()
def fibo(n):
    if n <= 2: 
        return 1
    else:
        return fibo(n-1) + fibo(n-2)
    
for i in range(1,10):
    print(fibo(i))
```

```python
from functools import cache

@cache
def fibo(n):
    if n <= 2: 
        return 1
    else:
        return fibo(n-1) + fibo(n-2)
    
for i in range(1,6):
    print(i, fibo(i))
```

## factorial using recursion

```python
import cProfile
def fact(n):
    if n == 0:
        return 1
    return n * fact(n-1)
# cProfile.run('fact(100)')
cProfile.run('fact(10), fact(20), fact(30)')
```

## factorial using memoization

```python
import cProfile
dic = {}
def fact(n):
    
    if n == 0:
        return 1
    elif n in dic:
        return dic[n]
    else:
        value = n * fact(n-1)
        dic[n] = value
        return value

# cProfile.run('fact(100)')
cProfile.run('fact(10), fact(20), fact(30)')
```

# Tree

```python
class BinaryTree(object):
    def __init__(self,rootObj):
        self.key = rootObj
        self.leftChild = None
        self.rightChild = None

    def insertLeft(self,newNode):
        if self.leftChild == None:
            self.leftChild = BinaryTree(newNode)
        else:
            t = BinaryTree(newNode)
            t.leftChild = self.leftChild
            self.leftChild = t

    def insertRight(self,newNode):
        if self.rightChild == None:
            self.rightChild = BinaryTree(newNode)
        else:
            t = BinaryTree(newNode)
            t.rightChild = self.rightChild
            self.rightChild = t


    def getRightChild(self):
        return self.rightChild

    def getLeftChild(self):
        return self.leftChild

    def setRootVal(self,obj):
        self.key = obj

    def getRootVal(self):
        return self.key
```

## Tree traversal
### preOrder

```python
def preOrder(tree):
    if tree:
        print(tree.getRootVal())
        preOrder(tree.getLeftChild())
        preOrder(tree.getRightChild())
preOrder(r)
```

### postOrder

```python
def postOrder(tree):
    if tree:
        postOrder(tree.getLeftChild())
        postOrder(tree.getRightChild())
        print(tree.getRootVal())
postOrder(r)
```

### inOrder

```python
def inOrder(tree):
    if tree:
        inOrder(tree.getLeftChild())
        print(tree.getRootVal())
        inOrder(tree.getRightChild())
inOrder(r)
```
